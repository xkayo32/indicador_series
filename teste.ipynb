```python
import pandas as pd
import numpy as np
```

```python
dataframe = pd.read_csv('previsao.csv')
```

```python
dataframe['retorno logaritmico'] = np.log(dataframe['Close'] / dataframe['Close'].shift(1))
```

```python
dataframe['tendencia logaritmico'] = np.log(
    dataframe['maxima_tendencia'] / dataframe['maxima_tendencia'].shift(1))
```

```python
dataframe['retorno logaritmico'].plot()
```

```python
dataframe['tendencia logaritmico'].plot()
```
